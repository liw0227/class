```python
# 读取 CSV 格式文件“task_2_lianjia_data.csv”，将数据命名为 lianjia_data，并选取前部数据查看(阅读 参考文章 1、2);
import pandas as pd
f = open('/Users/liwenjie/Downloads/task_2_lianjia_data.csv')
lianjia_data = pd.read_csv(f)
f.close
lianjia_data.head()
```

```python
# 利用 groupby 函数，以【面积大小(m²)】字段进行 groupby，统计各个面积下的【房源数量】情况(阅读参考 文章 3、4);
# lj_gb = lianjia_data.groupby('面积大小（㎡）').地铁站.agg({}'房源数量':'count'})  //旧写法
lj_gb = lianjia_data.groupby('面积大小（㎡）').地铁站.agg([('房源数量', 'count')])
lj_gb
```

```python
# 将第(2)问所统计数据，作折线图和条形图，并得出相关结论(阅读参考文章 5、6)，提示:先提取第(2)问 所得数据表的 index 索引，
# 饼转化成 list;再提取【房源数量】的值转成 list(阅读参考文章 7、8);

import matplotlib.pyplot as plt
x = lj_gb.index.values
y1 = lj_gb.values.tolist()
y = lj_gb['房源数量'].values.tolist()
print(y1)
print(y)
plt.plot(x,y,'bo-')
plt.bar(x,y, color='red',edgecolor='yellow')
plt.show()
```

```python
# 利用 groupby 函数，以【区】字段进行 groupby，统计各个上海各个“区”所属的【房源数量】情况;
area = lianjia_data.groupby('区').地铁站.agg([('房源数量','count')])
area
```

```python
# 将第(4)问所统计数据，作条形图，并得出相关结论;
'''显示中文'''
plt.rcParams['font.sans-serif']=['Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False
x1 = area.index.values
y1 = area['房源数量'].values.tolist()
print(x1,y1)
plt.bar(x1,y1,color = 'red')
plt.show()
```

```python
# 利用 groupby 函数，以【楼层类型】字段进行 groupby，统计各个楼层所属的【房源数量】情况;
floor_type = lianjia_data.groupby('楼层类型').地铁站.agg([('房源数量','count')])
floor_type
```

```python
# 将第(6)问所统计数据，作饼图(阅读参考文章 9)，并得出相关结论。
x2 = floor_type.index.values
y2 = floor_type['房源数量'].values.tolist()
print(x2,y2)
plt.pie(y2, labels=x2,autopct='%1.2f%%')
plt.show()
```
