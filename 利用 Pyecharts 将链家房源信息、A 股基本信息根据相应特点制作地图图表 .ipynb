```python
# 根据任务 4 所统计的上海各个区的房均价数据，利用 Pyecharts 中的 Map，以上海地图为背景作可视化展示图 (提示:可以直接定义 districts = ['金山区', '奉贤区', '嘉定区', '松江区', '青浦区', '宝山区', '闵行区 ', '浦东新区', '普陀区', '杨浦区', '虹口区', '长宁区', '静安区', '徐汇区', '黄浦区'])(阅读参考文章 2);
import tushare as ts
df = ts.get_stock_basics()
df
```

```python
import pandas as pd
f = open('/Users/liwenjie/Downloads/Task_4_lianjia_sale.csv')
f.close
df1 = pd.read_csv(f)
df1.head()
```

```python
avg = df1.groupby('区域').价格.agg([('各区平均价格','mean')]).sort_values(by='各区平均价格')
avg.round(2)
```

```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.charts import Bar
from pyecharts import options as opts
list_avg = avg['各区平均价格'].values.tolist()
area = avg.index.values.tolist()
bar = (
    Bar()
    .add_xaxis(area)
    .add_yaxis('房价',list_avg)
    .set_global_opts(title_opts=opts.TitleOpts(title="上海各区域平均价格"))
)
bar.reversal_axis()
bar.load_javascript() 
```

```python
bar.render_notebook()
```
